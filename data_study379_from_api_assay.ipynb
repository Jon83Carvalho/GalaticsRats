```python
#df_379_assay=pd.DataFrame(data_379_assay)
df_379_assay=pd.DataFrame(data_string.json())
```

```python
df_379_assay_table=pd.DataFrame.from_records(df_379_assay['tableData'][['current']][0])
```

```python
# Split the "Parameter Value: QA Score" column and extract the numerical value
df_379_assay_table['QA Score'] =  df_379_assay_table['Parameter Value[QA Score]'].str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
data_dict.to_json('data_379_assay.json')
```

```python
# Create a histogram using seaborn
plt.figure(figsize=(10, 6))
ax = sns.histplot(df_379_assay_table['QA Score'])

# Annotate each bar with its height
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.0f}', 
                xy=(p.get_x() + p.get_width() / 2, height), 
                xytext=(0, 5),  # 5 points vertical offset
                textcoords='offset points', 
                ha='center', va='bottom')

plt.title('Distribution of QA Score')
plt.xlabel('QA Score')
plt.ylabel('Frequency')
plt.show()
```
